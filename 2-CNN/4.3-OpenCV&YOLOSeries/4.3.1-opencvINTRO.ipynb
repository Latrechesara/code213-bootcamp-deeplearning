**What is OpenCV?**

OpenCV (Open Source Computer Vision Library) is an **open-source computer vision and machine learning library**.  
It is widely used for **image processing**, **video analysis**, and **real-time applications**.

Key reasons to learn OpenCV:  

- **Image & video processing**: load, manipulate, and display images or videos.  
- **Real-time capabilities**: supports live webcam or video stream processing.  
- **Integration**: works with Python, C++, and deep learning frameworks like PyTorch or TensorFlow.  
- **Rich functions**: filters, edge detection, transformations, object detection, and more.  

You can think of OpenCV as:  
- A **toolbox for computer vision** (like NumPy for arrays, but specialized for images/videos).  
- A **preprocessing library** for machine learning and deep learning pipelines.

## 4 - Loading & Displaying Images <a name="4"></a>

**Instruction:** In this exercise, you will learn how to load an image from your computer and display it using both OpenCV and Matplotlib.

**Steps:**  
1. Use `cv2.imread()` to load an image.  
2. Use `cv2.imshow()` to display the image in an OpenCV window.  
3. Use `matplotlib.pyplot.imshow()` to display the image inline in the notebook.  
4. Remember that OpenCV loads images in **BGR format**, so you may need to convert it to **RGB** before using Matplotlib.


````python
# Exercise: Load and display an image

# Step 1: Load the image (replace 'your_image.jpg' with your file)
image = cv2.imread('myimage.jpg')

# Step 2: Display using OpenCV
# Note: cv2.imshow may not work in Jupyter on some systems
# cv2.imshow('Image', image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

# Step 3: Convert to RGB for Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 4: Display inline using Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axes
plt.title("Loaded Image")
plt.show()
````

## 5 - Basic Image Operations <a name="5"></a>

### 5.1 - Resize & Crop <a name="5-1"></a>

**Instruction:** In this exercise, you will learn how to resize and crop images using OpenCV.

**Steps:**  
1. Resize the image to a new width and height using `cv2.resize()`.  
2. Crop a region of interest (ROI) from the image using array slicing.  
3. Display the resized and cropped images using Matplotlib.


````python
# Exercise: Resize and Crop an image

# Step 1: Resize the image to 300x300 pixels
resized_image = cv2.resize(image, (300, 300))

# Step 2: Crop a region of interest (e.g., top-left 100x100 pixels)
cropped_image = image[0:100, 0:100]  # [y1:y2, x1:x2]

# Step 3: Display the results
fig, axes = plt.subplots(1, 3, figsize=(12,4))

axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
axes[1].set_title('Resized Image')
axes[1].axis('off')

axes[2].imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('Cropped Image')
axes[2].axis('off')

plt.show()
````

### 5.2 - Convert to Grayscale <a name="5-2"></a>

**Instruction:** Convert your image to grayscale using OpenCV. This is often a preprocessing step for computer vision tasks like edge detection and object detection.

**Steps:**  
1. Use `cv2.cvtColor()` with `cv2.COLOR_BGR2GRAY`.  
2. Display the grayscale image using Matplotlib.


````python
# Exercise: Convert image to grayscale

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
````

## 6 - Drawing on Images <a name="6"></a>

**Instruction:** Learn to draw shapes and text on images using OpenCV. This is useful for annotations, highlighting regions, or marking objects in computer vision tasks.

**Steps:**  
1. Draw a **rectangle** around a region of interest.  
2. Draw a **circle** at a specific point.  
3. Draw a **line** between two points.  
4. Add **text** to the image using a chosen font and color.


````python
# Exercise: Draw shapes and text on an image

# Make a copy of the image to draw on
img_draw = image.copy()

# Step 1: Draw a rectangle (start_point, end_point, color, thickness)
cv2.rectangle(img_draw, (50, 50), (200, 200), (0, 255, 0), 3)

# Step 2: Draw a circle (center, radius, color, thickness)
cv2.circle(img_draw, (150, 150), 50, (255, 0, 0), 3)

# Step 3: Draw a line (start_point, end_point, color, thickness)
cv2.line(img_draw, (0, 0), (300, 300), (0, 0, 255), 2)

# Step 4: Add text (image, text, position, font, font_scale, color, thickness)
cv2.putText(img_draw, "OpenCV!", (50, 300), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2)

# Display the image
plt.imshow(cv2.cvtColor(img_draw, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Drawing Shapes and Text")
plt.show()
````

## 7 - Accessing Pixels <a name="7"></a>

**Instruction:** Learn how to access and modify pixel values. This is fundamental for image processing operations like filtering or masking.

**Steps:**  
1. Access the color of a specific pixel.  
2. Modify the color of a pixel or a region.  
3. Observe how changes affect the displayed image.


````python
# Exercise: Access and modify pixels

# Step 1: Access the color of pixel at (100, 100)
pixel_value = image[100, 100]
print("Pixel value at (100,100):", pixel_value)

# Step 2: Change the pixel color at (100, 100) to red
image[100, 100] = [0, 0, 255]

# Step 3: Modify a small square region to green
image[50:100, 50:100] = [0, 255, 0]

# Display the modified image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Pixel Access and Modification")
plt.show()
````

## 8 - Video Capture (Webcam) <a name="8"></a>

**Instruction:** Learn to capture video from your webcam using OpenCV and display it in real-time. This is a key skill for real-time computer vision applications like face detection, motion tracking, or object detection.

**Steps:**  
1. Initialize the webcam capture.  
2. Read frames in a loop.  
3. Display each frame using OpenCV.  
4. Add a way to **exit the loop** safely.  
5. Experiment with **frame processing**, e.g., converting to grayscale.


````python
# Exercise: Capture and display video from the webcam
# run this a stand alone python script, to avoid kernel crashes
# Step 1: Initialize webcam
import cv2

def main():
    # Initialize webcam
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        print("Cannot open webcam")
        return

    print("Press 'q' to quit")
    
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Optional: convert to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Show the frame
        cv2.imshow("Webcam - Press q to Quit", gray)

        # Exit when 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release resources
    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
````
